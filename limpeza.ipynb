Importação das bibliotecas

```python
import pandas as pd
```

Carregamento dos dados

```python
df_1 = pd.read_csv(r'dataset\VRA_20251.csv', skiprows=1, sep=';')
df_2 = pd.read_csv(r'dataset\VRA_20252.csv', skiprows=1, sep=';')
df_3 = pd.read_csv(r'dataset\VRA_20253.csv', skiprows=1, sep=';')
df_4 = pd.read_csv(r'dataset\VRA_20254.csv', skiprows=1, sep=';')
df_5 = pd.read_csv(r'dataset\VRA_20255.csv', skiprows=1, sep=';')
df_6 = pd.read_csv(r'dataset\VRA_20256.csv', skiprows=1, sep=';')
df_7 = pd.read_csv(r'dataset\VRA_20257.csv', skiprows=1, sep=';')
df_8 = pd.read_csv(r'dataset\VRA_20258.csv', skiprows=1, sep=';')
df_9 = pd.read_csv(r'dataset\VRA_20259.csv', skiprows=1, sep=';', low_memory=False)
df_10 = pd.read_csv(r'dataset\VRA_202510.csv', skiprows=1, sep=';')
```

Criação de um única df para limpar todos os arquivos

```python
dfs = [df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10]
df = pd.concat(dfs, ignore_index=True)
```

```python
print(f"Total de linhas: {df.shape[0]}")
df.head()
```

```python
df.info()
```

```python
df['Situação Voo'].value_counts()
```

```python
df['Código Tipo Linha'].value_counts()
```

```python
df['Código Autorização (DI)'].value_counts()
```

```python
df = df[df['Situação Voo'] == 'REALIZADO']
df.head()
```

```python
df = df.drop(columns=['Código Justificativa'])
```

```python
print(f"Total de linhas: {df.shape[0]}")
df.head()
```

```python
df['Código Tipo Linha'].value_counts()
```

```python
df = df[df['Código Tipo Linha'].isin(['N', 'I'])]
```

```python
df['Código Tipo Linha'].value_counts()
```

```python
df.shape
```

```python
df.info()
```

```python
colunas_data = ['Partida Prevista', 'Partida Real', 'Chegada Prevista', 'Chegada Real']

for coluna in colunas_data:
    df[coluna] = pd.to_datetime(df[coluna], errors='coerce', format='%d/%m/%Y %H:%M')    
```

```python
df.info()
```

```python
df['Partida Prevista'].isnull().sum()
```

```python
df = df.dropna(subset=['Partida Prevista'])
```

```python
df.info()
```

```python
df.drop(columns=['Situação Voo'], inplace=True)
```

```python
df.info()
```

```python
df = df.dropna(subset=['Chegada Prevista'])
```

```python
df.info()
```
